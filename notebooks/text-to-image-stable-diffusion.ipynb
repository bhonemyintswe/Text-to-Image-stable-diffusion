```python
from transformers import pipeline, set_seed
import warnings
warnings.filterwarnings("ignore")
```

```python
device = "cuda"
```

```python
import torch
from diffusers import StableDiffusionPipeline
from diffusers import DiffusionPipeline

# make sure you're logged in with `huggingface-cli login`
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16") 
pipe = pipe.to(device)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_image(prompet):
    image1 = pipe(prompt=prompet).images[0]
    plt.subplot(1, 1, 1)
    plt.imshow(image1)
    plt.title('stable-diffusion-xl-base-1.0')
    plt.axis('off')
```

```python
plot_image('Samurai warrior, ultra realistic, Black and Red, Japanese flag in background, portrait')
```

```python
plot_image('An illustration of a red owl with bright blue eyes in the style of abstract expressionism with volumetric lighting.')
```

```python
plot_image('Giant with blue marble skincolour, with cat pupils, visible sharp teeth, bodyshape like Dwayne Johnson, hairstyle like Khal Drogo, with glow - in - the - dark rune tattoo, leather trousers, near campfire at sunset, photorealistic fantasy style')
```

```python
plot_image('A retro-style robot playing a futuristic video game in neon tones with medium resolution')
```
